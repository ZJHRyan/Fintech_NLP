```python
import pandas as pd
import re
```

```python
sentence=[]
for line in open('text.txt', 'r').readlines():
    print(line.strip())
    sentence.append(re.findall('[\u4e00-\u9fa5]',line))
```

```python
sentence
```

```python
sen=pd.DataFrame(sentence)
```

```python
sen
```

```python
sen=sen.dropna(how='all')
```

```python
sen=sen.reset_index()
```
